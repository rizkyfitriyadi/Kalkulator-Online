# BASIC CALCULATOR

```python
def penjumlahan(angka_pertama,angka_kedua):
  hasil = angka_pertama + angka_kedua
  return hasil

def pengurangan(angka_pertama, angka_kedua):
  hasil = angka_pertama - angka_kedua
  return hasil

def perkalian(angka_pertama, angka_kedua):
  hasil = angka_pertama * angka_kedua
  return hasil

def pembagian(angka_pertama, angka_kedua):
  hasil = angka_pertama / angka_kedua
  return hasil
```

```python
while True:
  try:
    angka_pertama = int(input("Masukan angka pertama = "))
    angka_kedua = int(input("Masukan angka kedua = "))
    break
  except ValueError:
    print("Periksa kembali input kamu, pastikan hanya memasaukkan angka")

while True:
  pilih_operasi = input("Silahkan pilih operasi yang diinginkan (+|-|*|/) = ")
  pilihan_operasi =("+","-","*","/")
  if pilih_operasi in pilihan_operasi:
    if pilih_operasi == "+":
      hasil = penjumlahan(angka_pertama, angka_kedua)
    elif pilih_operasi == "-":
      hasil = pengurangan(angka_pertama, angka_kedua)
    elif pilih_operasi == "*":
      hasil = perkalian(angka_pertama, angka_kedua)
    elif pilih_operasi == "/":
      while angka_kedua == 0:
          print("Pembagian dengan nol tidak diperbolehkan, silakan masukkan angka kedua yang lain.")
          try:
              angka_kedua = int(input("Masukan angka kedua = "))
          except ValueError:
              print("Periksa kembali input kamu, pastikan hanya memasukkan angka")
              continue
      hasil = pembagian(angka_pertama, angka_kedua)
    print(f"Hasil dari {angka_pertama} {pilih_operasi} {angka_kedua} = {hasil} ")
  else:
    print("Periksa kembali input kamu, pastikan hanya memasaukan (+|-|*|/)")
    continue

  while True:
    operasi_lanjutan = input("Apakah kamu ingin melakukan operasi lagi? (ya/tidak) = ")
    if operasi_lanjutan.lower() == "ya":
      angka_pertama = hasil
      try:
        angka_kedua = int(input("Masukan angka kedua = "))
        break
      except ValueError:
        print("Periksa kembali input kamu, pastikan hanya memasaukkan angka")
        continue
    elif operasi_lanjutan.lower() == "tidak":
      print("Terima kasih telah menggunakan program kalkulator by Rizky Fitriyadi, Semoga harimu menyenangkan!")
      break
    else:
      print("Periksa kembali input kamu, pastikan hanya memasaukkan (ya/tidak)")
      continue
  if operasi_lanjutan.lower() == "tidak":
    break
```

    <script>
        // State management
        let currentOperation = '';
        let lastResult = null;
        let isChaining = false;

        // DOM elements
        const display = document.getElementById('display');
        const angka1Input = document.getElementById('angka1');
        const angka2Input = document.getElementById('angka2');
        const calculateBtn = document.getElementById('calculateBtn');
        const continueSection = document.getElementById('continueSection');
        const continueYes = document.getElementById('continueYes');
        const continueNo = document.getElementById('continueNo');
        const resetBtn = document.getElementById('resetBtn');
        const operationBtns = document.querySelectorAll('.operation-btn');

        // Error elements
        const error1 = document.getElementById('error1');
        const error2 = document.getElementById('error2');
        const errorOp = document.getElementById('errorOp');

        // Operation functions
        function penjumlahan(a, b) {
            return a + b;
        }

        function pengurangan(a, b) {
            return a - b;
        }

        function perkalian(a, b) {
            return a * b;
        }

        function pembagian(a, b) {
            if (b === 0) {
                throw new Error('Pembagian dengan nol tidak diperbolehkan');
            }
            return a / b;
        }

        // Clear errors
        function clearErrors() {
            error1.textContent = '';
            error2.textContent = '';
            errorOp.textContent = '';
        }

        // Validate inputs
        function validateInputs() {
            clearErrors();
            let isValid = true;

            const angka1 = parseFloat(angka1Input.value);
            const angka2 = parseFloat(angka2Input.value);

            if (isNaN(angka1) || angka1Input.value === '') {
                error1.textContent = 'Periksa kembali input kamu, pastikan hanya memasukkan angka';
                isValid = false;
            }

            if (isNaN(angka2) || angka2Input.value === '') {
                error2.textContent = 'Periksa kembali input kamu, pastikan hanya memasukkan angka';
                isValid = false;
            }

            if (!currentOperation) {
                errorOp.textContent = 'Periksa kembali input kamu, pastikan memilih operasi (+|-|*|/)';
                isValid = false;
            }

            return isValid;
        }

        // Operation button handlers
        operationBtns.forEach(btn => {
            btn.addEventListener('click', () => {
                // Remove active class from all buttons
                operationBtns.forEach(b => b.classList.remove('active'));
                
                // Add active class to clicked button
                btn.classList.add('active');
                
                currentOperation = btn.dataset.op;
                clearErrors();
            });
        });

        // Calculate function
        function calculate() {
            if (!validateInputs()) {
                return;
            }

            const angka1 = parseFloat(angka1Input.value);
            const angka2 = parseFloat(angka2Input.value);
            let hasil;

            try {
                switch (currentOperation) {
                    case '+':
                        hasil = penjumlahan(angka1, angka2);
                        break;
                    case '-':
                        hasil = pengurangan(angka1, angka2);
                        break;
                    case '*':
                        hasil = perkalian(angka1, angka2);
                        break;
                    case '/':
                        hasil = pembagian(angka1, angka2);
                        break;
                    default:
                        throw new Error('Operasi tidak valid');
                }

                // Display result
                const operationSymbol = currentOperation === '*' ? '×' : currentOperation === '/' ? '÷' : currentOperation;
                display.innerHTML = `
                    <div style="font-size: 18px; margin-bottom: 10px;">
                        ${angka1} ${operationSymbol} ${angka2} =
                    </div>
                    <div style="font-size: 32px; color: #48bb78;">
                        ${hasil}
                    </div>
                `;

                lastResult = hasil;
                continueSection.classList.add('show');

            } catch (error) {
                if (error.message.includes('nol')) {
                    error2.textContent = error.message;
                } else {
                    display.innerHTML = `<div style="color: #e53e3e;">Error: ${error.message}</div>`;
                }
            }
        }

        // Continue with calculation
        function continueCal() {
            angka1Input.value = lastResult;
            angka2Input.value = '';
            angka2Input.focus();
            continueSection.classList.remove('show');
            
            // Clear operation selection
            operationBtns.forEach(b => b.classList.remove('active'));
            currentOperation = '';
            
            display.innerHTML = `
                <div style="font-size: 18px; margin-bottom: 10px;">
                    Lanjutan dari hasil:
                </div>
                <div style="font-size: 28px; color: #667eea;">
                    ${lastResult}
                </div>
            `;
            isChaining = true;
        }

        // Reset calculator
        function resetCalculator() {
            angka1Input.value = '';
            angka2Input.value = '';
            currentOperation = '';
            lastResult = null;
            isChaining = false;
            
            operationBtns.forEach(b => b.classList.remove('active'));
            continueSection.classList.remove('show');
            clearErrors();
            
            display.textContent = 'Masukkan angka untuk memulai';
            angka1Input.focus();
        }

        // Event listeners
        calculateBtn.addEventListener('click', calculate);
        continueYes.addEventListener('click', continueCal);
        continueNo.addEventListener('click', () => {
            display.innerHTML = `
                <div style="font-size: 18px; color: #48bb78;">
                    Terima kasih telah menggunakan program kalkulator by Rizky Fitriyadi
                </div>
                <div style="font-size: 16px; margin-top: 10px; color: #666;">
                    Semoga harimu menyenangkan! 😊
                </div>
            `;
            continueSection.classList.remove('show');
        });
        resetBtn.addEventListener('click', resetCalculator);

        // Enter key support
        document.addEventListener('keypress', (e) => {
            if (e.key === 'Enter') {
                if (continueSection.classList.contains('show')) {
                    continueCal();
                } else {
                    calculate();
                }
            }
        });

        // Input validation on type
        [angka1Input, angka2Input].forEach(input => {
            input.addEventListener('input', clearErrors);
        });

        // Focus on first input when page loads
        window.addEventListener('load', () => {
            angka1Input.focus();
        });
    </script>
